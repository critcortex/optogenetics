```python
import numpy as np
import pylab
%pylab inline           
```

```python
xs = np.arange(-1.0,10.,0.01)

def log(xs,x0,a=1.,b=1.,c=0.):
    """
    Of the form: a*np.log(b*xs+x0)+c
    Params:
        xs
        x0
        a
        b
        c
    """
    ys = np.zeros(xs.size)
    ys[xs>x0-1] = a*np.log(b*xs+1)+c
    ys[ys<c] = c
    return ys


pylab.figure()
ys1 = log(xs,2,c=2)
ys2 = log(xs,0)
pylab.plot(xs,ys1,'r')
pylab.plot(xs,ys2,'b')
ylim(ymin=-1)
#xlim(xs[0],xs[-1])

```

```python


def sigmoid(x,k,x0,y0=0.,c=1.):
    """
    Of the form: c / (1 + np.exp(-k*(x-x0))) + y0
    
    Params:
        x
        k
        x0
        y0 
        c
    """
    return c / (1 + np.exp(-k*(x-x0))) + y0

def product(xs,a,log_x0,sig_k,sig_x0,w=0.7):
    return a*((w*log(xs,log_x0))*((1-w)*sigmoid(xs,sig_k,sig_x0)))

def summed(xs,a,log_x0,sig_k,sig_x0,w=0.7):
    return a*((w*log(xs,log_x0))+((1-w)*sigmoid(xs,sig_k,sig_x0)))

def linear_sigmoid(xs,gradient_coeff,sigmoid_coeff):
    tmp = sigmoid_coeff*sigmoid(xs,1,1) + gradient_coeff*xs
    return tmp

def combo(xs,gradient_coeff,prod_coeff,a,log_x0,sig_k,sig_x0,gradient_x0,w1=0.9):
    return prod_coeff*product(xs,a,log_x0,sig_k,sig_x0,w1)+ gradient_coeff*(xs-gradient_x0)

def heaviside(xs):
    xs[xs<0] = 0
    return xs
```
