```python
import numpy as np

def plot_photocurrents(tree,vm_index=1,irr=0.01):
    nb,ns,nl = tree
    print nb,ns,nl
    pylab.figure()
    fname = '../140305_compare_equal_branchesirr%.2f_factor0.00_nb%g_ns%g_nl%g_highExp3_NpHR_none_ChR_whole_v.dat'%(irr,nb,ns,nl)
    print fname
    data = np.loadtxt(fname)
    tm = data[0,:]
    vm = data[vm_index,:]    
    pylab.plot(tm,vm,'b')
    for f in np.arange(0.1,1.01,0.1):
        try:
            data = np.loadtxt('../140305_compare_equal_branchesirr%.2f_factor%.2f_nb%g_ns%g_nl%g_highExp3_NpHR_whole_ChR_whole_v.dat'%(irr,f,nb,ns,nl))
            tm = data[0,:]
            vm = data[vm_index,:]
            pylab.plot(tm,vm,'b')
        except:
            continue
            
    for f in [2.,5.,10.]:
        try:
            data = np.loadtxt('../140305_compare_equal_branchesirr%.2f_factor%.2f_nb%g_ns%g_nl%g_highExp3_NpHR_whole_ChR_whole_v.dat'%(irr,f,nb,ns,nl))
            tm = data[0,:]
            vm = data[vm_index,:]
            pylab.plot(tm,vm,'b')    
        except:
            continue

    pylab.ylim(-150,0)
    pylab.savefig('compare_photocurrents_tree%g_%g_%g_irr%.2f_recLoc%g.png'%(nb,ns,nl,irr,vm_index))
    return pylab.gcf()
```
