## Plot cells

```python
import LFPy
import Neuron
import os
from matplotlib.collections import PolyCollection
import matplotlib.pyplot as plt
import pylab
%pylab inline
```

```python

cell = LFPy.Cell(morphology='../morphologies/cell1.asc')
rotation = {'x' : 1.233, 'y' : 0, 'z' : 0}
cell.set_rotation(**rotation)
zips = []
for x, z in cell.get_idx_polygons():
    zips.append(zip(x, z))

polycol = PolyCollection(zips,
                         linewidths=1,
                         edgecolors='black',
                         facecolors='black')

fig = pylab.figure(figsize=(5,15))
ax = fig.add_subplot(111)

ax.add_collection(polycol)
ax.axis(ax.axis('equal'))

#pylab.show()
pylab.savefig('Fig2A_plotL5PC.png',dpi=300)
```
