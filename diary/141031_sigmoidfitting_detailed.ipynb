```python
def get_optdescript(irr,factor):
    opt_des = []
    if irr > 0 :
        opt_des.append('whole')
    else:
        opt_des.append('none')
        
    if irr > 0 and factor > 0:
        opt_des.append('whole')
    else:
        opt_des.append('none')
    return opt_des
```
