```python
import NeuroTools.stgen as ntst 
import numpy as np
import pylab
import math
%pylab inline
```

```python
stg = ntst.StGen()
tt = np.arange(0,1000)
tta = np.arange(0,1001)
psth = np.zeros_like(tt)
T = 2*math.pi
gamma = 40
w = 1000./gamma/T
for i in xrange(1000):
    ss = stg.inh_poisson_generator(1000*np.sin(1./w*tt)+0,tt,1000,array=True)
    #print ss[:10]
    psth+=np.histogram(ss,bins=1000)[0]

# normalize

psth = psth.astype(float)

pylab.plot(tt,psth,linestyle='steps')
pylab.xlim(0,100)
pylab.show()
```

```python
bg_r = 70
high_r = 140
for i in range(500):
    pp = stg.inh_poisson_generator((high_r-bg_r)*np.sin(1./w*tt)+bg_r,tt,1000,array=True)
    if i<10:
        print len(pp[np.where(pp<1000)])
    ys = i*np.ones(len(pp))
    pylab.scatter(pp,ys,marker='.')

pylab.xlim(0,100)
```
